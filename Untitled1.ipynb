```python
import pandas as pd

df = pd.read_csv("uniprot_to_pdb_id.csv")
uniprots = list(df["From"])
pdb_ids = list(df["To"])

uniprot_to_pdb_id_map = {}

for u, p in zip(uniprots, pdb_ids):
    uniprot_to_pdb_id_map[u] = p
    
# print(len(uniprot_to_pdb_id_map))
for p_id in uniprot_to_pdb_id_map.values():
    print(p_id)
```
