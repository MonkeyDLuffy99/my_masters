```python
import pandas as pd
import numpy as np
from scipy.linalg import fractional_matrix_power
import matplotlib.pyplot as plt

from pysmiles import read_smiles
import networkx as nx
from tqdm import tqdm as tqdm
import warnings
warnings.filterwarnings("ignore")

'''
Graph Convolutional Neural Network implementation without training.
Based on the implementation of Inneke Mayachita, 
https://towardsdatascience.com/understanding-graph-convolutional-networks-for-node-classification-a2bfdb7aba7b
'''

MATRIX_SIZE = 50 # variability in matrix sizes

df = pd.read_csv("protac.csv")
protacs = df["Smiles"]

molecule_to_atomic_number = {
    "C":6,
    "O":8,
    "N":7,
    "S":16,
    "Cl":17,
    "F":7,
    "Br":35,
    "I":53,
    "P":15
}

protac_adjacency_matrices = []
graph_features = []
for i in range(len(protacs)): 
    mol = read_smiles(protacs[i])
    elements = nx.get_node_attributes(mol, name = "element")
    X = []
    for val in elements.values():
        X.append(molecule_to_atomic_number[val])
    X_ = np.array(X)
    X_.resize(MATRIX_SIZE)
    graph_features.append(X_)
    A = nx.adjacency_matrix(mol, weight='order').todense() 
    Adj = np.array(A)
    Adj.resize(MATRIX_SIZE*MATRIX_SIZE)
    Adj = np.reshape(Adj, (MATRIX_SIZE, MATRIX_SIZE))
    protac_adjacency_matrices.append(Adj)
```

```python
protac_adjacency_matrices = np.array(protac_adjacency_matrices)
graph_features = np.array(graph_features)
print(protac_adjacency_matrices.shape)
print(protac_adjacency_matrices[0].shape)
print(graph_features.shape)
print(graph_features[0].shape)
```

```python
from tqdm import tqdm as tqdm

np.random.seed(77777)
hidden_neurons = MATRIX_SIZE
output_size = 120
Weight_0 = np.random.randn(MATRIX_SIZE, hidden_neurons) * 0.05
Weight_1 = np.random.randn(hidden_neurons, output_size) * 0.05

# ReLU
def relu(x):
    return np.maximum(0, x)

def gcn(A, H, W):
    graph = nx.from_numpy_matrix(A)
    self_loops = []
    for i in range(graph.number_of_nodes()):
        self_loops.append((i, i))

    graph.add_edges_from(self_loops)
    A_with_self_loops = nx.to_numpy_array(graph)
    D = np.diag(np.sum(A_with_self_loops, axis=1))  
    D_half_norm = fractional_matrix_power(D, -0.5)
    result = D_half_norm.dot(A_with_self_loops).dot(D_half_norm).dot(H).dot(W)
    return relu(result)

def forward_pass(matrix, features, first_weight, second_weight):
    H1 = gcn(matrix, features, first_weight)
    H2 = gcn(matrix, H1, second_weight)
    
    return H2

data = []
for i in tqdm(range(len(graph_features))):
    example = forward_pass(protac_adjacency_matrices[i], graph_features[i], Weight_0, Weight_1)
    data.append(example)

data = np.array(data)
```
